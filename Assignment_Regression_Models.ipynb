````python
# put it into a pandas dataframe

df = pd.DataFrame(boston['data'])

df
````

#### Multiple Linear Regression
The exact same thing, but with multiple variables...

```
Y = b0 + b1X1 + b2X2....

Listing Price of House = $100,000 + ($100 * Square Footage) + ($10,000 * Number of Bedrooms)
```

````python
corr = df.corr()

corr.style.background_gradient(cmap='coolwarm')
````

````python
vars = [
        'RM', 'LSTAT','PTRATIO',  # high +- corr
        'DIS','RAD'
        ]

X = df[vars]
y = df.target

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.75, random_state=2)
````

````python
X_train
````

````python
y_test
````

````python
multi_lr = LinearRegression()
````

````python
multi_lr.fit(X_train, y_train)
````

````python
multi_lr.intercept_
````

````python
multi_lr.coef_
````

````python
multi_predictions = multi_lr.predict(X_test)
````

````python
y_test.tail()
````

````python
r2_score(y_true=y_test, y_pred=multi_predictions)
````

````python
mean_absolute_error(y_true=y_test, y_pred=multi_predictions)
````
